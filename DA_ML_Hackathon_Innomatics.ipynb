### Haversine distance formula

```python
import math
 
def haversine(lat1, lon1, lat2, lon2):
 
    dLat = (lat2 - lat1) * math.pi / 180.0
    dLon = (lon2 - lon1) * math.pi / 180.0
 
    lat1 = (lat1) * math.pi / 180.0
    lat2 = (lat2) * math.pi / 180.0
 
    # apply formulae
    a = (pow(math.sin(dLat / 2), 2) +
         pow(math.sin(dLon / 2), 2) *
             math.cos(lat1) * math.cos(lat2));
    rad_earth = 6371
    c = 2 * math.asin(math.sqrt(a))
    return rad_earth * c
```

```python
df_temp = df[['pickup_latitude','pickup_longitude','dropoff_latitude','dropoff_longitude']]
```

```python
df_temp
```

```python
df['haversine_distance'] = df_temp.apply(lambda x: haversine(x.pickup_latitude,x.pickup_longitude,x.dropoff_latitude,x.dropoff_longitude),axis = 1)
```

#### Haversine distance calculates the spherical distance between two points. IN the above case where the fare_amount is maximum but the distance is minimum could be because the rider must have started from the pick up location and dropped at the same location after taking his/her journey.

#### Model buuilding and finding the best algorithm

```python
df['ride_week_day'] = df['pickup_datetime'].dt.weekday
```

```python
X,y = df[['passenger_count', 'haversine_distance', 'ride_week_day']],df['fare_amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size = 0.3,
                                                   random_state = 42)
```

#### Linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
reg_model = LinearRegression()
reg_model.fit(X_train,y_train)
y_predicted = reg_model.predict(X_test)
r2_linear = r2_score(y_test,y_predicted)
print(r2_linear)
```

#### Decision Tree Classifier won't work for our data as target is continuous. We go with Decision Tree Regressor

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dregressor = DecisionTreeRegressor()
dregressor.fit(X_train, y_train)
y_pred = dregressor.predict(X_test)
r2_dt = r2_score(y_test,y_pred)
print(r2_dt)
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfregressor = RandomForestRegressor(n_estimators = 100)
rfregressor.fit(X_train, y_train)
y_pred = rfregressor.predict(X_test)
r2_rf = r2_score(y_test,y_pred)
print(r2_rf)
```

#### KNN Classifier won't work for our data as target is continuous. We go with KNN regressor.

#### KNN Regressor

```python
from sklearn import neighbors
# Considering number of neighbours as 10
KNNmodel = neighbors.KNeighborsRegressor(n_neighbors = 10)
KNNmodel.fit(X_train, y_train)
y_pred=KNNmodel.predict(X_test)
r2_knn = r2_score(y_test,y_pred)
print(r2_knn)
```
